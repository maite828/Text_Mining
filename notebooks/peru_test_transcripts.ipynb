```python
import re
import os
import codecs

output = {}

regex_pseudo = '\s+([^\W\d_]+)\s+\((.*?)\)'
regex_sampa = '\s+([^\W\d_]+)\s+\[(.*?)\]'
regex_words = '\s+([^\W\d_]+)\s+'

full_ark = '/Users/zoraida/Desktop/test_peru_ark/byturn/708001035011988.ark'
ark = '708001035011988.ark'

with codecs.open(full_ark, encoding="utf-8") as f:
    byturn = f.readline()

    # Process pseudofonectic 
    pseudofonectic = re.findall(regex_pseudo, byturn, re.UNICODE)
    for orto, fone in pseudofonectic:   
        fone = fone.replace(' ', '_')
        if not fone in output:
            output[fone] = [orto, [ark], 1]
        else:
            output[fone][2] = output[fone][2] + 1
            output[fone][1].append(ark)
    byturn = re.sub(regex_pseudo, '', byturn, count=0)

    # Process sampa
    sampa_trans = re.findall(regex_sampa, byturn, re.UNICODE)
    for orto, sampa in sampa_trans:  
        sampa = sampa.replace(' ', '_')
        if not sampa in output:
            output[sampa] = [orto, [ark], 1]
        else:
            output[sampa][2] = output[sampa][2] + 1
            output[sampa][1].append(ark)
    byturn = re.sub(regex_sampa, '', byturn, count=0)

    # Process words
    words = re.findall(regex_words, byturn, re.UNICODE)
    for w in words:
        if not w in output:
            output[w] = [w, [ark], 1]
        else:
            output[w][2] = output[w][2] + 1 
            output[w][1].append(ark)
            
with codecs.open('vocabulary_freq.txt', 'w', encoding="utf-8") as f:
    for key, value in sorted(output.items(), key=lambda e: e[1][0]):
        f.write('\t'.join([key] + [value[0]] + [str(value[2])] ) + '\n')
        
with codecs.open('vocabulary_ark.txt', 'w', encoding="utf-8") as f:
    for key, value in sorted(output.items(), key=lambda e: e[1][0]):
        f.write(key + '\t' + ','.join(value[1]) + '\n')
```

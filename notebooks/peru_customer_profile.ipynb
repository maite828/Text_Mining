```python
# We group by ani, pla_id_anexo and we sum the values of the rest of the attributes(for the 199 skill categories, how many times the ani called)
calls_df_tmp_sum = calls_df_tmp.groupby(['ANI', 'PLA_ID_ANEXO']).sum().reset_index()
```
